# Векторные представления текстов

В этой тетрадке я векторизую (tf-idf) 4 русскоязычных текста из Википедии и сравниваю их между собой с помощью косинусой близости.

Тексты:
- `football.txt` — статья про футбол;
- `hockey.txt` — статья про хоккей;
- `machine_learning.txt` — статья про машинное обучение;
- `neural_network.txt` — статья про нейронную сеть.

## Импорт библиотек

Для решения я использую TF-IDF-векторизацию. Каждый текст представляется как разреженный вектор: ключи — слова и биграммы, значения — их TF-IDF-веса.

```python
import math
import re
from collections import Counter
from pathlib import Path
```

## Чтение текстов

```python
texts_dir = Path("texts")
text_paths = sorted(texts_dir.glob("*.txt"))

names = []
texts = []

for path in text_paths:
    text = path.read_text(encoding="utf-8")
    names.append(path.name)
    texts.append(text)

names
```

## Токенизация

Я беру не только отдельные слова, но и биграммы, то есть пары соседних слов. Например, для технических текстов это помогает сохранить выражения вроде `машинное обучение` и `нейронная сеть`.

```python
word_pattern = re.compile(r"[^\W\d_]{2,}", re.UNICODE)

tokenized_texts = []

for text in texts:
    words = word_pattern.findall(text.lower())
    bigrams = []

    for i in range(len(words) - 1):
        bigrams.append(words[i] + " " + words[i + 1])

    tokenized_texts.append(words + bigrams)

for name, tokens in zip(names, tokenized_texts):
    print(name, len(tokens))
```

## Словарь и частоты документов

Для IDF нужно знать, в скольких текстах встретился каждый токен.

```python
vocabulary = set()

for tokens in tokenized_texts:
    vocabulary.update(tokens)

vocabulary = sorted(vocabulary)

document_frequencies = Counter()

for tokens in tokenized_texts:
    unique_tokens = set(tokens)

    for token in unique_tokens:
        document_frequencies[token] += 1

len(vocabulary)
```

## TF-IDF-векторизация

```python
vectors = []
total_documents = len(texts)

for tokens in tokenized_texts:
    token_counts = Counter(tokens)
    total_tokens = len(tokens)
    vector = {}

    for token, count in token_counts.items():
        tf = count / total_tokens
        idf = math.log((1 + total_documents) / (1 + document_frequencies[token])) + 1
        vector[token] = tf * idf

    vectors.append(vector)

len(vectors)
```

## Cosine similarity

Чем ближе значение к 1, тем больше похожи тексты по набору слов и биграмм.

```python
similarities = []

for first_vector in vectors:
    row = []

    for second_vector in vectors:
        common_tokens = first_vector.keys() & second_vector.keys()

        dot_product = 0
        for token in common_tokens:
            dot_product += first_vector[token] * second_vector[token]

        first_norm = math.sqrt(sum(value ** 2 for value in first_vector.values()))
        second_norm = math.sqrt(sum(value ** 2 for value in second_vector.values()))

        if first_norm == 0 or second_norm == 0:
            similarity = 0
        else:
            similarity = dot_product / (first_norm * second_norm)

        row.append(similarity)

    similarities.append(row)
```

## Результат

```python
best_i = 0
best_j = 1
best_score = similarities[best_i][best_j]

for i in range(len(names)):
    for j in range(i + 1, len(names)):
        if similarities[i][j] > best_score:
            best_i = i
            best_j = j
            best_score = similarities[i][j]

print(f"Файлы: {', '.join(names)}")
print(f"Размер словаря: {len(vocabulary)}")
print()
print("Матрица cosine similarity:")

short_names = [Path(name).stem for name in names]
column_width = max(12, max(len(name) for name in short_names) + 2)

print(" " * column_width + "".join(f"{name:>{column_width}}" for name in short_names))

for name, row in zip(short_names, similarities):
    scores = "".join(f"{score:>{column_width}.3f}" for score in row)
    print(f"{name:<{column_width}}{scores}")

print()
print(f"Самые близкие тексты: {names[best_i]} и {names[best_j]} ({best_score:.3f})")
```

## Комментарий к результату

Перед подсчетом ожидалось, что тексты разобьются на две тематические пары: `machine_learning.txt` будет близок к `neural_network.txt`, а `football.txt` - к `hockey.txt`. В целом это ожидание подтвердилось: технические тексты действительно оказались заметно ближе друг к другу, чем к спортивным, а спортивные тексты - ближе друг к другу, чем к техническим.

Самой близкой парой оказались `football.txt` и `hockey.txt` со значением cosine similarity около 0.364. Пара `machine_learning.txt` и `neural_network.txt` тоже близкая, но ее значение ниже - около 0.290. Это получилось потому, что TF-IDF сравнивает тексты по пересечению слов и биграмм. В статьях про футбол и хоккей много общей спортивной лексики: `игра`, `команда`, `игрок`, `матч`, `ворота`, `чемпионат`, `правила`, `соревнования`. В технических статьях тоже есть общие слова, например `обучение`, `модель`, `данные`, `сеть`, но статья про нейронные сети содержит много специфических терминов, которых меньше в статье про машинное обучение. Поэтому спортивная пара получилась самой близкой.
